```julia
paths = readdir(joinpath(expanduser("~"),
                "Datasets/ecommerce_product_images/Footwear/Men/Images/images_with_product_ids"), 
                join=true);
```

```julia
im1 = load(paths[1])
```

```julia
images = [load(p) for p in paths];
```

```julia
avg_hashes = [average_hash(im) for im in images]
dif_hashes = [difference_hash(im) for im in images];
```

```julia
function pairwise_distances(images, hashes)
    n_images = length(images)
    D = zeros(Int16, n_images, n_images)

    for i in 1:n_images
        for j in 1:n_images
            D[i,j] = hamming_bitwise(hashes[i], hashes[j])
        end
    end
    return D
end
```

```julia
D = pairwise_distances(images, dif_hashes);
```

```julia
q = 5
topk = 10
distances = D[q,:]
sorted_indices = sortperm(distances);
top_K = sorted_indices[1:topk];

print("Query image is images[$q] corresponding to $(split(paths[5],'/')[end])")
imresize(images[q],(256,256))
```
